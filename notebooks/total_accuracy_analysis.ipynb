```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
df_wd = pd.read_csv("./audio_accuracy_wd.csv")
df = pd.read_csv("./audio_accuracy_no_wd.csv")
```

```python
df = df.sort_values(by=['accuracy', 'accuracy_top5'], ascending=False)
def rename(x):
    return os.path.basename(x)
df["file_name"] = df["file_name"].apply(rename)

df_wd = df_wd.sort_values(by=['accuracy', 'accuracy_top5'], ascending=False)
df_wd["file_name"] = df_wd["file_name"].apply(rename)

display(df)
```

```python
values = df["accuracy"].tolist()
values_wd = df_wd["accuracy"].tolist()
x = np.linspace(0, len(values), len(values))
plt.figure(figsize=(8,5))
plt.title("TOP-1 Accuracy per song")
# plt.plot(x, values)
plt.plot(x, values_wd)
```

```python
df = df.sort_values(by=['accuracy_top5', 'accuracy'], ascending=False)
df_wd = df_wd.sort_values(by=['accuracy_top5', 'accuracy'], ascending=False)
values = df["accuracy_top5"].tolist()
values_wd = df_wd["accuracy_top5"].tolist()
x = np.linspace(0, len(values), len(values))
plt.plot(x, values)
plt.plot(x, values_wd)
display(df_wd)
```
